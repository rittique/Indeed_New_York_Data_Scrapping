```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import datetime

def scrape_indeed_jobs(skill, location, num_pages):
    # set up the driver
    chrome_options = Options()

    PROXY = "" # HOST:PORT

    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument(f'--proxy-server={PROXY}')

    driver = webdriver.Chrome(executable_path=r'chromedriver', options=chrome_options)
    
    # create empty lists to store job details
    job_titles = []
    company_names = []
    locations = []
    salaries = []
    job_types = []
    schedules = []
    date_posted = []
    current_date = []
    job_descriptions = []
    apply_links = []
    
    # loop through the number of pages and scrape each page
    for page in range(num_pages):
        # get the url for the page
        url = f"https://www.indeed.com/jobs?q={skill}&l={location}&sort=date&start={page*10}"
        driver.get(url)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # get the job details from the page
        for job in soup.find_all('div', 'slider_container css-77eoo7 eu4oa1w0'):
            # job title
            title = job.find('a').text.strip()
            job_titles.append(title)
            
            # company name
            company = job.find('span', {'class': 'companyName'}).text.strip()
            company_names.append(company)
            
            # location
            loc = job.find('div', {'class': 'companyLocation'}).text.strip()
            locations.append(loc)
            """
            # salary
            salary = job.find('span', {'class': 'salaryText'})
            if salary:
                salary = salary.text.strip()
            else:
                salary = 'Not specified'
            salaries.append(salary)
            
            
            
            try:
                # job type and schedule
                job_type = job.find('span', {'class': 'jobType'})
                job_type = job_type.text.strip()
            except:
                job_type = 'Not specified'
            
            job_types.append(job_type)
            
            """
            
            # date posted
            date = job.find('span', {'class': 'date'}).text.strip()
            date_posted.append(date)
            
            # apply link
            apply_link = 'https://www.indeed.com/' + job.h2.a.get('href')
            apply_links.append(apply_link)
            
            # current date
            curr_date = datetime.datetime.now().strftime("%Y-%m-%d")
            current_date.append(curr_date)
            
            # job description
            driver.get(apply_link)
            desc_soup = BeautifulSoup(driver.page_source, 'html.parser')
            desc = desc_soup.find('div', {'class': 'jobsearch-jobDescriptionText'})
            if desc:
                job_descriptions.append(desc.text.strip())
            else:
                job_descriptions.append('Not specified')
            
            
    
    # create a dataframe from the job details
    df = pd.DataFrame({
        'Job Title': job_titles,
        'Company Name': company_names,
        'Location': locations,
        'Salary': salaries,
        'Job Type': job_types,
        'Schedule': schedules,
        'Date Posted': date_posted,
        'Current Date': current_date,
        'Job Description': job_descriptions,
        'Apply Link': apply_links
    })
    
    driver.quit()
    return df
```
