```python
directory = os.listdir(r"C:\Users\Rittique\Python\Indeed_Newyork_Webscrapping\Excel_data_workbooks")
```

```python
final_list = []
for file in directory:
    if ".csv" in file:
        final_list.append(file)
final_list
```

```python
dfs = []

final_df = pd.DataFrame()
for f in final_list:
    df = pd.read_csv("./Excel_data_workbooks/"+f)
    final_df = pd.concat([final_df, df], axis=0)
    
final_df
```

```python
final_df = final_df.reset_index(level=None, drop=True, inplace=False)
final_df
```

```python
final_df = final_df.drop("Unnamed: 0", axis='columns')
final_df
```
